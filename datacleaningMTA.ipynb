```python
import pandas as pd


mta = pd.read_csv("MTA_Subway_Stations_20260402.csv")
```

```python
cols_keep = [
    "GTFS Stop ID",
    "Complex ID",
    "Stop Name",
    "Borough",
    "Daytime Routes",
    "Structure",
    "GTFS Latitude",
    "GTFS Longitude",
    "ADA"
]
mta_filtered = mta[cols_keep].copy()
```

```python
print(mta_filtered.shape)
print(mta_filtered.head())
print(mta_filtered.info())
```

```python
print(mta_filtered.isna().sum())
```

```python
print(mta_filtered["Borough"].value_counts(dropna=False))
print(mta_filtered["Structure"].value_counts(dropna=False))
print(mta_filtered["ADA"].value_counts(dropna=False))
```

```python
print(mta_filtered[["GTFS Latitude", "GTFS Longitude"]].describe())
```

## EDA


```python
mta_filtered["route_count"] = mta_filtered["Daytime Routes"].astype(str).str.split().str.len()
# see if it is a big station 
```

```python
print(mta_filtered[["Daytime Routes", "route_count"]].head())
print(mta_filtered["route_count"].describe())
```

```python
print("GTFS Stop ID duplicates:", mta_filtered["GTFS Stop ID"].duplicated().sum())
print("Complex ID duplicates:", mta_filtered["Complex ID"].duplicated().sum())
print("Stop Name duplicates:", mta_filtered["Stop Name"].duplicated().sum())
```

```python
mta_distance = mta_filtered.copy()
mta_map = mta_filtered.drop_duplicates(subset=["Complex ID"]).copy()
```

```python
mta_distance.to_csv("MTA_Subway_Stations_distance.csv", index=False)
mta_map.to_csv("MTA_Subway_Stations_map.csv", index=False)
```
