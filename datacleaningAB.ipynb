## Data cleaning Prising


```python
print((ab["price"] == 0).sum())
print(ab["price"].sort_values(ascending=False).head(200))
```

```python
print(ab["price"].quantile([0.5, 0.75, 0.9, 0.95, 0.99]))
```

```python
ab_filtered = ab_filtered[ab_filtered["minimum_nights"] <= 30]
print(ab_filtered["minimum_nights"].describe())
```

## LAT,LON CHECKING

```python
print(ab_filtered[["latitude", "longitude"]].describe())
```

```python
ab["last_review"] = pd.to_datetime(ab["last_review"], errors="coerce")
print(ab["last_review"].min(), ab["last_review"].max())
```

## missing data

```python
ab_filtered["name"] = ab_filtered["name"].fillna("Unknown")
ab_filtered["host_name"] = ab_filtered["host_name"].fillna("Unknown")

ab_filtered["last_review"] = pd.to_datetime(ab_filtered["last_review"], errors="coerce")

ab_filtered["reviews_per_month"] = ab_filtered["reviews_per_month"].fillna(0)
```

## last check ab_filtered

```python
print(ab_filtered.isna().sum())
```

```python
print(ab_filtered["price"].describe())
print((ab_filtered["price"] == 0).sum())
print(ab_filtered["price"].sort_values(ascending=False).head(10))
```

```python
print(ab_filtered["minimum_nights"].describe())
print(ab_filtered["minimum_nights"].sort_values(ascending=False).head(10))
print(ab_filtered[["latitude", "longitude"]].describe())
```

```python
def classify_price(group):
    q25 = group["price"].quantile(0.25)
    q75 = group["price"].quantile(0.75)

    def label_price(price):
        if price < q25:
            return "Low"
        elif price <= q75:
            return "Medium"
        else:
            return "High"

    group["price_level_in_hood"] = group["price"].apply(label_price)
    return group

ab_filtered = ab_filtered.groupby("neighbourhood_group", group_keys=False).apply(classify_price)
```

```python
ab_filtered.to_csv("AB_NYC_2019_cleaned.csv", index=False)
```
